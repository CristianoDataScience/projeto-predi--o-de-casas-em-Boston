### Modelo de Regressão com SRV 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold, cross_val_score
```

```python
df = pd.read_csv("housing.csv", sep=",", encoding="iso-8859-1")
df.head(2)
```

```python
independente = df.iloc[:, 0:3].values
independente
```

```python
independente.shape
```

```python
dependente = df.iloc[:, 3].values
dependente
```

```python
dependente.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(independente, dependente, test_size=0.3, random_state=0)
```

```python
x_train.shape, x_test.shape
```

```python
modelo_svr = SVR(kernel='poly')
modelo_svr.fit(x_train, y_train)
```

```python
modelo_svr.score(x_train, y_train)
```

```python
modelo_svr.score(x_test, y_test)
```

```python
previsoes = modelo_svr.predict(x_train)
previsoes
```

```python
x_scaler = StandardScaler()
x_train_scaler = x_scaler.fit_transform(x_train)
```

```python
x_train_scaler
```

```python
y_scaler = StandardScaler()
y_train_scaler = y_scaler.fit_transform(y_train.reshape(-1, 1))
```

```python
y_train_scaler
```

```python
y_scaler = StandardScaler()
y_train_scaler = y_scaler.fit_transform(y_train.reshape(-1, 1))
```

```python
y_train_scaler
```

```python
x_test_scaler = x_scaler.transform(x_test)
x_test_scaler
```

```python
y_test_scaler = y_scaler.transform(y_test.reshape(-1, 1))
y_test_scaler
```

```python
modelo_svr2 = SVR(kernel='rbf')
modelo_svr2.fit(x_train_scaler, y_train_scaler.ravel()) # .ravel() é para retornar matriz 1D
```

```python
modelo_svr2.score(x_train_scaler, y_train_scaler)
```

#### Teste

```python
modelo_svr2.score(x_test_scaler, y_test_scaler)
```

```python
previsoes_teste = modelo_svr2.predict(x_test_scaler)
previsoes_teste
```

```python
y_test_scaler
```

### Revertendo a transformação

```python
y_test_inverse = y_scaler.inverse_transform(y_test_scaler)
previsoes_inverse = y_scaler.inverse_transform(previsoes_teste.reshape(-1, 1))
```

```python
y_test_inverse
```

```python
previsoes_inverse
```

```python
mean_absolute_error(y_test_inverse, previsoes_inverse)
```

### Métricas

```python
# Erro médio Absoluto
mean_absolute_error(y_test_inverse, previsoes_inverse)
```

```python
# Erro quadrático médio
mean_squared_error(y_test_inverse, previsoes_inverse)
```

```python
# Raiz do erro quadrático médio (RMSE)
np.sqrt(mean_squared_error(y_test_inverse, previsoes_inverse))
```
